```python
#Node Class
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next

#Link List Class
class linklist:
    def __init__(self):
        self.head=None
        
    def at_beigning(self,data):
        node=Node(data,self.head)
        self.head=node
        
    def Print(self):
        if self.head is None:
            print('Empty Linked List')
            return
        itr=self.head
        llstr=''
        while itr:
            llstr+=str(itr.data)+'-->'
            itr=itr.next
        print(llstr)
    
    def at_end(self,data):
        if self.head is None:
            at_beigning(data,None)
            return
        itr=self.head
        while itr.next:
            itr=itr.next
        itr.next=Node(data)
        
    def ll_len(self):
        if self.head is None:
            print('Linked List is Empty')
            return
        itr=self.head
        count=0
        while itr:
            itr=itr.next
            count+=1
        return count
    
    def insert_at(self,index,data):
        if index < 0 or index >= self.ll_len():
            raise Exception('Invalid Index')
        if index == 0:
            self.at_beigning(data)
        
        itr=self.head
        count=0
        while itr:
            if index-1 == count:
                node=Node(data,itr.next)
                itr.next=node
            itr=itr.next
            count+=1
            
    def delete_at(self,index):
        if index < 0 or index >= self.ll_len():
            raise Exception('Invalid Index')
        itr=self.head
        if index == 0:
            self.head=itr.next
        count=0
        while itr:
            if index-1 == count:
                itr.next=itr.next.next
            itr=itr.next
            count+=1
            
    def Search_by_index(self,index):
        if index < 0 or index >= self.ll_len():
            raise Exception('Invalid Index')
            
        itr=self.head
        count=0
        while itr:
            if index == count:
                return itr.data
            itr=itr.next
            count+=1
    
    def Search_by_value(self,data):
        itr=self.head
        count=0
        while itr:
            if itr.data==data:
                return count
            itr=itr.next
            count+=1
        return f'{data} is Not in Linked List'
    
    def del_by_value(self,data):
        itr=self.head
        count=0
        while itr:
            if itr.data == data:
                self.delete_at(count)
            itr=itr.next
            count+=1
            
    def insert_at_mid(self,data):
        mid=(self.ll_len()//2)
        itr=self.head
        count=0
        while itr:
            if mid-1 == count:
                node=Node(data,itr.next)
                itr.next=node
            itr=itr.next
            count+=1
            
    def insert_after_value(self,data,inst):
        itr=self.head
        count=0
        while itr:
            if itr.data == data:
                self.insert_at(count+1,inst)
            itr=itr.next
            count+=1

if __name__=='__main__':
    ll=linklist()
    ll.Print()
    ll.at_beigning(6)
    ll.at_beigning(63)
    ll.at_beigning(16)
    ll.at_end(12)
    ll.at_end(7)
    ll.Print()
    ll.insert_at(4,5)
    ll.Print()
    ll.delete_at(4)
    ll.Print()
    print(ll.Search_by_index(4))
    print(ll.Search_by_value(90))
    ll.del_by_value(63)
    ll.Print()
    ll.insert_at_mid(9)
    ll.Print()
    ll.insert_at_mid(11)
    ll.Print()
    ll.insert_after_value(11,15)
    ll.Print()
    print(ll.ll_len())
```
