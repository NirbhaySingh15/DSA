```python
#Queue usinf list method
class Queue:
    
    def __init__(self):
        self.size=int(input('Enter the size of Queue: '))
        self.queue=[]
        
    def Insert(self):
        if self.Length() >= self.size:
            print('\nQUEUE IS FULL')
        else:
            data=int(input('\nEnter the Value: '))
            self.queue.append(data)
    
    def Delete(self):
        if self.Length() == 0:
            print('\nQUEUE IS EMPTY')
        else:
            self.queue.pop(0)
    
    def Length(self):
        length=len(self.queue)
        return length
    
    def Print(self):
        if self.Length() == 0:
            print('\nQUEUE IS EMPTY')
        else:
            print(end='QUEUE: ')
            for i in self.queue:
                print(i,end=' ')
    
    def Clear(self):
        if self.Length() == 0:
            print('\nQUEUE IS ALREADY EMPTY')
        else:
            self.queue.clear()
            
if __name__ == "__main__":
    qu=Queue()
    
    loop = -1
    while loop == -1:
        select=int(input('\nSelect the Action:\n1.Insert 2.Delete 3.Print 4.Ele_len 5.Clear 6.Exit: '))
        if select == 1:
            qu.Insert()
        elif select == 2:
            qu.Delete()
        elif select == 3:
            qu.Print()
        elif select == 4:
            print(qu.Length())
        elif select == 5:
            qu.Clear()
        elif select == 6:
            loop=1
            print('\nThankyou!!!')
        else:
            print('\nINVALID CREDENTIAL')
```
