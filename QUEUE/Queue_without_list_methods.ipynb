```python
class Queue:
    def __init__(self):
        self.size=int(input('\nEnter the size of Queue: '))
        self.queue=[]
    
    def Insert(self):
        if self.Length() >= self.size:
            print('\nQUEUE IS FULL')
        else:
            data=int(input('\nEnter the value: '))
            self.queue[self.Length():]=[data]
    
    def Delete(self):
        if self.Length() == 0:
            print('\nQUEUE IS EMPTY')
        else:
            self.queue=self.queue[1:]
    
    def Length(self):
        count=0
        for i in self.queue:
            count+=1
        return count
    
    def Print(self):
        if self.Length() == 0:
            print('\nQUEUE IS EMPTY')
        else:
            print('\nQueue: ',end='')
            for i in self.queue:
                print(i,end=' ')
            print('\n')
        
    def Clear(self):
        if self.Length() == 0:
            print('\nQUEUE IS ALREADY EMPTY')
        else:
            self.queue=[]
            
if __name__ == "__main__":
    
    qu=Queue()
    
    loop=-1
    while loop==-1:
        select=int(input('\nSelect the Action:\n1.Insert 2.Delete 3.Print 4.Ele_len 5.Clear 6.Exit: '))
        if select == 1:
            qu.Insert()
        elif select == 2:
            qu.Delete()
        elif select == 3:
            qu.Print()
        elif select == 4:
            print('Element in Queue: ',qu.Length())
        elif select == 5:
            qu.Clear()
        elif select == 6:
            loop=1
            print('\nThankyou!!!')
        else:
            print('\nINVALID CREDENTIAL')
```
