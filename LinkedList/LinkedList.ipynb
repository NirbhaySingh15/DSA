```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next
        
class linkedList:
    def __init__(self):
        self.head=None
        
    def atBegning(self,n):
        node=Node(n,self.head)
        self.head=node
        
    def atEnd(self,n):
        if self.head is None:
            print('Linked List is Empty!!!')
            return
        itr=self.head
        while itr.next:
            itr=itr.next
        itr.next=Node(n)
        
    def atIndex(self,data,n):
        if n < 0 or n >= self.Length():
            raise Exception('Invalid Index!!!')
        if n==0:
            self.head=Node(data,self.head)
        itr=self.head
        count=0
        while itr:
            if n-1 == count:
                itr.next=Node(data,itr.next)
            itr=itr.next
            count+=1
            
    def insetAfterVAlue(self,data,n):
        itr=self.head
        while itr:
            if itr.data == n:
                itr.next=Node(data,itr.next)
                return
            itr=itr.next
        print(f'{n} is not in Linked List')
            
    def delByIndex(self,n):
        if n < 0 or n >= self.Length():
            raise Exception('Invalid Index!!!')
        if n == 0:
            self.head=self.head.next
        itr=self.head
        count=0
        while itr:
            if n-1==count:
                itr.next=itr.next.next
            itr=itr.next
            count+=1
        
    def delAtBeign(self):
        if self.head is None:
            print('Linked List is Empty!!!')
            return
        self.head=self.head.next
        
    def delAtEnd(self):
        if self.head is None:
            print('Linked List is Empty!!!')
            return
        itr=self.head
        while itr.next.next:
            itr=itr.next
        itr.next=None
        
    def delAfterValue(self,n):
        itr=self.head
        while itr:
            if itr.data == n:
                itr.next=itr.next.next
            itr=itr.next
        print(f'{n} is not in Linked List')
        
    def searchByIndex(self,n):
        if n < 0 or n >= self.Length():
            print('Invalid Index!!!')
            return
        itr=self.head
        count=0
        while itr:
            if count == n:
                print(f'index number {n} have {itr.data}')
                return
            itr=itr.next
            count+=1
            
    def searchByValue(self,n):
        itr=self.head
        count=0
        while itr:
            if itr.data == n:
                print(f'{n} is on {count} index')
                return
            itr=itr.next
            count+=1
        print(f'{n} is not in Linked List')
        return
    
    def midElement(self):
        if self.head is None:
            print('Linked List is Empty!!!')
            return
        if self.Length()==0:
            print(f'Middle Element if {self.head.data}')
        mid=self.Length()//2
        itr=self.head        
        count=0
        while itr:
            if count == mid:
                print(f'Middle Element is {itr.data}')
                return
            itr=itr.next
            count+=1

    def Length(self):
        if self.head is None:
            print('Linked List is Empty!!!')
            return
        itr=self.head
        count=0
        while itr:
            itr=itr.next
            count+=1
        return count
        
    def Print(self):
        if self.head is None:
            print('Linked List is Empty!!!')
            return
        itr=self.head
        ll=''
        while itr:
            ll+=str(itr.data)+'->'
            itr=itr.next
        print(ll)
    
if __name__=='__main__':
    # 13 Functions
    LL=linkedList()
    LL.atBegning(5)
    LL.Print()
    LL.midElement()
    LL.atBegning(7)
    LL.Print()
    LL.midElement()
    LL.atBegning(4)
    LL.Print()
    LL.midElement()
    LL.atEnd(9)
    LL.atEnd(3)
    LL.atEnd(1)
    LL.Print()
    LL.delAtEnd()
    LL.Print()
    LL.delAtBeign()
    LL.Print()
    print('Linked List: ',LL.Length())
    LL.atIndex(10,2)
    LL.atIndex(11,0)
    LL.atIndex(12,2)
    LL.Print()
    LL.delByIndex(0)
    LL.Print()
    LL.delByIndex(3)
    LL.Print()
    print('Linked List: ',LL.Length())
    LL.searchByIndex(2)
    LL.searchByValue(9)
    LL.midElement()
    LL.insetAfterVAlue(11,5)
    LL.Print()
    LL.delAfterValue(11)
    LL.Print()
```
