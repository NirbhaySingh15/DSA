```python
class Node:
    def __init__(self,data=None,next=None,prev=None):
        self.data=data
        self.next=next
        self.prev=prev
        
class DoublyLinkedList:
    def __init__(self):
        self.head=None
        
    def insertAtBegning(self,data):
        if self.head is None:
            node=Node(data,self.head,None)
            self.head=node
        else:
            node=Node(data,self.head,None)
            self.head.prev=node
            self.head=node
            
    def insertAtEnd(self,data):
        if self.head is None:
            self.head=Node(data,self.head,None)
            return
        itr=self.head
        while itr.next:
            itr=itr.next
        itr.next=Node(data,None,itr)
    
    def insertAtIndex(self,data,n):
        if n < 0 or n > self.Length():
            print('Invalid Index!!!')
            return
        if n == 0:
            self.insertAtBegning(data)
            return
        itr=self.head
        count=0
        while itr:
            if n-1 == count:
                node=Node(data,itr.next,itr)
                if itr.next:
                    itr.next.prev=node
                itr.next=node
                break
            itr=itr.next
            count+=1
            
    def insertAfterValue(self,data,n):
        itr=self.head
        while itr:
            if itr.data == n:
                node=Node(data,itr.next,itr)
                if itr.next:
                    itr.next.prev=node
                itr.next=node
                return
            itr=itr.next
        print(f'{n} is not in Doubly Linked List!!!')
                   
    def deleteFromBegning(self):
        if self.head is None:
            print('Empty Doubly Linked List')
            return
        if self.Length()==1:
            self.head=None
        else:
            self.head=self.head.next
            self.head.prev=None
    
    def deleteFormEnd(self):
        if self.head is None:
            print('Empty Linked List!!!')
            return
        if self.Length()==1:
            self.head=None
        else:
            itr=self.head
            while itr.next.next:
                itr=itr.next
            itr.next=None
            
    def deleteByIndex(self,n):
        if n < 0 or n >= self.Length():
            print('Invalid Index!!!')
            return
        if self.Length()==1:
            self.head=None
        else:
            itr=self.head
            count=0
            while itr:
                if n-1 == count:
                    itr.next.next.prev=itr
                    itr.next=itr.next.next
                    break
                itr=itr.next
                count+=1
                
    def deleteByValue(self,n):
        itr=self.head
        while itr:
            if itr.next.data == n:
                itr.next.prev=itr
                itr.next=itr.next.next
                return
            itr=itr.next
        print(f'{n} is not in Doubly Linked List')
        
    def deleteAfterVAlue(self,n):
        if n == self.Last_Element().data:
            print(f'No value after {n}')
            return
        itr=self.head
        while itr:
            if itr.data == n:
                itr.next.prev=itr
                itr.next=itr.next.next
                return
            itr=itr.next
        print(f'{n} is not in Doubly Linked List!!!')
        
    def searchByIndex(self,n):
        if n < 0 or n >= self.Length():
            print('Invalid Index!!!')
            return
        itr=self.head
        count=0
        while itr:
            if count == n:
                print(f'Index {n} have {itr.data}')
                return
            itr=itr.next
            count+=1
            
    def searchByValue(self,n):
        itr=self.head
        count=0
        while itr:
            if itr.data == n:
                print(f'{n} is on {count} index')
                return
            itr=itr.next
            count+=1
        print(f'{n} is not in Doubly Linked List')
    
    def Print(self):
        if self.head is None:
            print('Empty Doubly Linked List!!!')
            return
        itr=self.head
        dll=''
        while itr:
            dll+=str(itr.data)+'->'
            itr=itr.next
        print(dll)
        
    def Reverse_print(self):
        if self.head is None:
            print('Empty Doubly Linked List!!!')
            return
        itr=self.Last_Element()
        rdll=''
        while itr:
            rdll+=str(itr.data)+'<-'
            itr=itr.prev
        print(rdll)
        
    def Length(self):
        if self.head is None:
            print('Empty Linked List!!!')
            return
        itr=self.head
        count=0
        while itr:
            itr=itr.next
            count+=1
        return count
        
    def Last_Element(self):
        itr=self.head
        while itr.next:
            itr=itr.next
        return itr
        
if __name__=='__main__':
    DLL=DoublyLinkedList()
    DLL.insertAtBegning(8)
    DLL.insertAtBegning(4)
    DLL.insertAtBegning(3)
    DLL.insertAtBegning(9)
    DLL.Print()
    DLL.Reverse_print()
    DLL.insertAtEnd(14)
    DLL.insertAtEnd(23)
    DLL.insertAtEnd(77)
    DLL.Print()
    DLL.Reverse_print()
    print("Doubly Linked List's Length:",DLL.Length())
    DLL.deleteFromBegning()
    DLL.Print()
    DLL.Reverse_print()
    DLL.deleteFormEnd()
    DLL.Print()
    DLL.Reverse_print()
    DLL.insertAtIndex(12,2)
    DLL.Print()
    DLL.Reverse_print()
    DLL.insertAfterValue(54,114)
    DLL.Print()
    DLL.deleteByIndex(2)
    DLL.Print()
    DLL.Reverse_print()
    DLL.deleteByValue(23)
    DLL.Print()
    DLL.Reverse_print()
    DLL.deleteAfterVAlue(8)
    DLL.Print()
    DLL.Reverse_print()
    DLL.searchByIndex(2)
    DLL.searchByValue(4)
    DLL.searchByValue(40)
    DLL.Reverse_print()
```
