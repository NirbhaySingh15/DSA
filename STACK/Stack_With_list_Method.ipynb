```python
#stack class with list method
class STACK:
    
    #construction function
    def __init__(self):
        self.size=int(input('Enter the size: '))
        self.stack=[]
    
    #Method for inserting element from the end
    def Insert(self):
        if len(self.stack) >= self.size:
            print('\nSTACK IS FULL')
        else:
            data=int(input('\nEnter the data: '))
            self.stack.append(data)
    
    #Method for deleting element from the end
    def Delete(self):
        if len(self.stack) == 0:
            print('\nSTACK IS EMPTY')
        else:
            self.stack.pop()
            
    #print the element of stack
    def Print(self):
        if len(self.stack) == 0:
            print('\nSTACK IS EMPTY')
        else:
            print('\nStack: ',end=' ')
            for i in self.stack:
                print(i,end=' ')
            
    #Number of element in stack 
    def Length(self):
        print(len(self.stack))
    
    #Deleting all Element from stack
    def Clean(self):
        if len(self.stack) == 0:
            print('\nSTACK IS ALREADY EMPTY')
        else:
            self.stack.clear()
            print('\nSTACK IS EMPTY')

if __name__ == '__main__':
    
    #Class Object
    st=STACK()
    
    #Option Loop
    loop=-1
    while loop == -1:
        select=int(input('\nPlease Select the Action:\n1.Insert 2.Delete 3.Print 4.Num_ele 5.Clear 6.Exit:  '))
        if select == 1:
            st.Insert()
        elif select == 2:
            st.Delete()
        elif select == 3:
            st.Print()
        elif select == 4:
            st.Length()
        elif select == 5:
            st.Clean()
        elif select == 6:
            loop = 1
            print('\nThankYou!!!!')
        else:
            print('\nINVALID CREDENTIAL')
```
